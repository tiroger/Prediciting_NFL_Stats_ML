```python
# Dependencies
import requests
import json
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
df_stadium = pd.read_excel('stadiums_nfl.xlsx',dtype = {"ZIP": str})
df_stadium
```

```python
df_stadium_zip = df_stadium["ZIP"].tolist()
df_stadium_zip
```

```python
df_start_date = pd.read_excel('date.xlsx')
df_start_date['Start_Date'] = pd.to_datetime(df_start_date.Start_Date)
df_start_date['Start_Date'] = df_start_date['Start_Date'].dt.strftime('%Y-%m-%d')
df_start_date = df_start_date["Start_Date"].tolist()
df_start_date
```

```python
def get_hourly_weather_for_zip(zip_code):

    data = []

    # Target city
    for date in df_start_date:
        start_date = date
    
        # Build the endpoint URL
        target_url = ('http://api.worldweatheronline.com/premium/v1/past-weather.ashx?key=255ee87dca7441b08fc180253192909&q={}&format=json&date={}&enddate=2018-12-31').format(zip_code,start_date)
        weather = requests.get(target_url).json()
        weather_list = weather['data']['weather']
        for weather_date in weather_list:
            d = weather_date['date']
            snow_cm = weather_date['totalSnow_cm']
            hourly_list = weather_date['hourly']
            for hour in hourly_list:
                temp_f = hour['tempF']
                time = hour['time']
                desc = hour['weatherDesc'][0]['value']
                humidity = hour['humidity']
                vis = hour['visibilityMiles']
                windspeedKmph = hour['windspeedKmph']
                precipInches=hour['precipInches']          

                row = {'zip_code': zip_code, 'date': d, 'snow':snow_cm,'time': time, 'temperature': temp_f,'windspeed':windspeedKmph,'precipitation':precipInches ,'visibility': vis,'humidity': humidity ,'desc': desc}
                data.append(row)   
    return data
```

```python
zipcode_data = []

for zipcode in df_stadium_zip:
    combined = get_hourly_weather_for_zip(zipcode)
    zipcode_data.extend(combined)
```

```python
df_final = pd.DataFrame.from_records(zipcode_data)
df_final = df_final[['zip_code', 'date', 'time','desc', 'temperature','snow','windspeed','visibility','humidity','precipitation']]
df_final_1500 = df_final.loc[df_final['time'] == '1500']
df_final_1500
```

```python
df_final_1500.to_csv('weather_final.csv')
```
